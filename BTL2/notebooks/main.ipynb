##Mục tiêu

1.  **EDA (Exploratory Data Analysis):**
    *   Thống kê mô tả dữ liệu, trực quan hóa để hiểu rõ các biến.
    *   Phát hiện dữ liệu thiếu, giá trị ngoại lai và xu hướng phân phối.

2.  **Tiền xử lý dữ liệu (Data Preprocessing):**
    *   Làm sạch dữ liệu, xử lý các giá trị bị thiếu.

    *    Hiện đại (deep learning embeddings): sử dụng các mô hình học sâu để biểu diễn văn bản.
4.  **Huấn luyện mô hình (Model Training):**Mô hình phân
 loại (Naive Bayes, Logistic Regression, SVM, hoặc fine-tuning mô hình BERT).

5.  **Đánh giá mô hình (Model Evaluation):**Dựa trên các thông số accuracy, F1-score.

##2.4 Stopwords & ký tự đặc biệt

### Tỉ lệ StopWorlds

```python
def stopword_ratio(text):
    words = str(text).split()
    if len(words) == 0:
        return 0
    return sum(1 for w in words if w.lower() in STOPWORDS) / len(words)

df["stopword_ratio"] = df["text"].apply(stopword_ratio)
print("📊 Tỷ lệ stopwords trung bình:", df["stopword_ratio"].mean())
```

### Thống kê ký tự đặc biệt

```python
df["num_punct"] = df["text"].apply(lambda x: sum(1 for c in str(x) if c in string.punctuation))
df["num_digits"] = df["text"].apply(lambda x: sum(1 for c in str(x) if c.isdigit()))
df["num_urls"] = df["text"].apply(lambda x: len(re.findall(r"http[s]?://\S+", str(x))))

print("\n🔹 Trung bình số dấu câu:", df["num_punct"].mean())
print("🔹 Trung bình số chữ số:", df["num_digits"].mean())
print("🔹 Trung bình số URL:", df["num_urls"].mean())
```

**Nhận xét:**

 Tỷ lệ stopwords trung bình khá cao (gần 40%), việc loại bỏ stopwords là một bước tiền xử lý hợp lý. Dữ liệu cũng chứa nhiều dấu câu và chữ số, cần được xử lý tùy thuộc vào yêu cầu của từng phương pháp embedding và mô hình.

### Tiền xử lý dữ liệu

```python
def clean_text(text):
    # Lowercase
    if CONFIG["preprocessing"]["lowercase"]:
        text = text.lower()

    # Remove content in brackets []
    if CONFIG["preprocessing"]["remove_brackets"]:
        text = re.sub(r'\[.*?\]', '', text)

    # Remove URLs
    if CONFIG["preprocessing"]["remove_urls"]:
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    if CONFIG["preprocessing"]["remove_html"]:
        text = re.sub(r'<.*?>+', '', text)

    # Remove punctuation
    if CONFIG["preprocessing"]["remove_punct"]:
        text = text.translate(str.maketrans("", "", string.punctuation))

    # Remove newline
    if CONFIG["preprocessing"]["remove_newline"]:
        text = text.replace("\n", " ")

    # Remove words with numbers
    if CONFIG["preprocessing"]["remove_numbers"]:
        text = re.sub(r'\w*\d\w*', '', text)

    # Remove non-word characters (symbols)
    if CONFIG["preprocessing"]["remove_symbols"]:
        text = re.sub(r"\\W", " ", text)

    # Remove stopwords
    if CONFIG["preprocessing"]["remove_stopwords"]:
        text = " ".join([w for w in text.split() if w not in STOPWORDS])


    tokens = None
    # Tokenization
    if CONFIG["preprocessing"]["tokenization"] == "word":
        tokens = word_tokenize(text)
        if CONFIG["preprocessing"]["remove_stopwords"]:
            tokens = [w for w in tokens if w not in STOPWORDS]

    elif CONFIG["preprocessing"]["tokenization"] == "bert":
        tokens = bert_tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=CONFIG["preprocessing"]["max_len"],
            truncation=True,
            padding="max_length" if CONFIG["preprocessing"]["padding"] else False,
            return_tensors="pt"
        )
        return tokens  # BERT trả về input_ids, attention_mask

    # Padding cho word tokenization
    if CONFIG["preprocessing"]["padding"] and tokens is not None:
        max_len = CONFIG["preprocessing"]["padding_length"]
        if len(tokens) < max_len:
            tokens = tokens + ["<PAD>"] * (max_len - len(tokens))
        else:
            tokens = tokens[:max_len]

    return tokens if tokens is not None else text
```

#3.Trích xuất đặc trưng

**Trích xuất đặc trưng**

```python
if CONFIG["embedding"] == "bow":
    vectorizer = CountVectorizer()
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)

elif CONFIG["embedding"] == "tfidf":
    vectorizer = TfidfVectorizer()
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
elif CONFIG["embedding"] == "word2vec":
    from gensim.models import Word2Vec
    import numpy as np

    # tiền xử lý: chuyển text thành list từ
    X_train_tok = [str(x).lower().split() for x in X_train]
    X_test_tok  = [str(x).lower().split() for x in X_test]

    # huấn luyện Word2Vec trên dữ liệu train
    w2v_model = Word2Vec(sentences=X_train_tok, vector_size=100, window=5, min_count=1, workers=4)

    def get_w2v_embeddings(tokenized_texts, model, vector_size=100):
        all_embeddings = []
        for tokens in tokenized_texts:
            vectors = [model.wv[w] for w in tokens if w in model.wv]
            if len(vectors) == 0:
                all_embeddings.append(np.zeros(vector_size))
            else:
                all_embeddings.append(np.mean(vectors, axis=0))
        return np.vstack(all_embeddings)

    X_train_vec = get_w2v_embeddings(X_train_tok, w2v_model)
    X_test_vec  = get_w2v_embeddings(X_test_tok, w2v_model)

else:
    raise ValueError("❌ Embedding không hợp lệ!") # option: bow | tfidf | word2Vec
```

##3. Mạng nơ-ron học đặc trưng – BERT Fine-tuning


Subtask:
Token hóa và mã hóa dữ liệu văn bản sử dụng `BertTokenizer`, tạo các input tensor (input_ids, attention_mask, token_type_ids) phù hợp với định dạng đầu vào của BERT, và chia dữ liệu thành tập train, validation và test.


**Reasoning**:
Tokenize and encode the text data using BertTokenizer, create input tensors, and split the data into training, validation, and test sets according to the instructions.



```python
# Load tokenizer và model BERT
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# Tokenization
def tokenize_fn(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=128)
```

**Tạo dataset HuggingFace**

```python
train_ds = Dataset.from_dict({"text": X_train.tolist(), "label": y_train.tolist()}).map(tokenize_fn, batched=True)
val_ds   = Dataset.from_dict({"text": X_val.tolist(),   "label": y_val.tolist()}).map(tokenize_fn, batched=True)
test_ds  = Dataset.from_dict({"text": X_test.tolist(),  "label": y_test.tolist()}).map(tokenize_fn, batched=True)

for ds in [train_ds, val_ds, test_ds]:
    ds.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
```

Ở đây, BERT chính là mô hình học đặc trưng tự động (representation learning).
Không cần thiết kế thủ công các đặc trưng ngữ pháp hay thống kê.

##4. Huấn luyện đầu-cuối (End-to-End Training)

```python
# Định nghĩa hàm metric
def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='macro')
    return {"accuracy": acc, "f1": f1}

# Tham số huấn luyện
training_args = TrainingArguments(
    output_dir="./bert_results",
    eval_strategy="epoch", # Changed from evaluation_strategy
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```
